### Membuat Dict

```python
data = {
  'total_records': 0,
  'attributes': [],
  'object_attributes': 0,
  'number_attributes': 0,
  'null_attributes': 0
}

data['total_records'] = df.shape[0]
data['attributes'] = list(df.columns.values)
data['null_attributes'] = list(df.loc[:, df.isnull().any()].columns)
data['object_attributes'] = list(df.select_dtypes(['object']).columns)
data['number_attributes'] = list(df.select_dtypes(['float64', 'int64']).columns)
```

### Menampilkan Data

```python
table = PrettyTable(["Informasi", "Jumlah"])
table.align = "l"

table.add_row(["Record", data['total_records']])
table.add_row(["Atribut", len(data['attributes'])])
table.add_row(["Atribut Object", len(data['object_attributes'])])
table.add_row(["Atribut Angka", len(data['number_attributes'])])
table.add_row(["Atribut Null", len(data['null_attributes'])])

print(table)

table = PrettyTable(["No.", "Nama Atribut"])
table.align = "l"
for i in range(20) :
    table.add_row([i, data['attributes'][i]])
    
print(table)

table = PrettyTable(["No.", "Nama Atribut Object"])
table.align = "l"
for i in range(6) :
    table.add_row([i, data['object_attributes'][i]])
    
print(table)
                           
table = PrettyTable(["No.", "Nama Atribut Angka"])
table.align = "l"
for i in range(17) :
    table.add_row([i, data['number_attributes'][i]])
    
print(table)
                           
table = PrettyTable(["No.", "Nama Atribut Nilai Null"])
table.align = "l"
for i in range(4) :
    table.add_row([i, data['null_attributes'][i]])
    
print(table)
                          
```
