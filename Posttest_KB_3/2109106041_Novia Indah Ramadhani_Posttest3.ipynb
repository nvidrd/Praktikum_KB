# Visualisasi
## ----------------------------------------------------------------
## 1. Grafik Pie

```python
data_pie = df['education'].value_counts().rename('education').reset_index(name = 'id_count')
data_pie.head()
```

```python
plt.pie(data_pie.id_count,
        labels     = data_pie['index'],
        startangle = 170,
        autopct    = '%.1f%%',
        colors     = ['#F9C5D5', '#F999B7', '#F2789F'],
        explode    = (0.1, 0.1, 0.1),
        shadow     = True,
       )

plt.title('Grafik Pie Dari Gelar Pendidikan Karyawan', fontsize = 15)
plt.show()
```

### Kesimpulan :
### Kesimpulan yang dapat diambil dari grafik di atas adalah pendidikan karyawan mayoritas bergelar Bachelor's dengan presentase sebanyak 70,0%. Pada urutan kedua terdapat karyawan yang bergelar Master's & above dengan presentase sebanyak 28,5%. Pada urutan ketiga terdapat karyawan yang bergelar Below Secondary dengan presentase sebanyak 1,5%.

## ----------------------------------------------------------------
## 2. Grafik Batang

```python
# Dengan Seaborn
plt.figure(figsize=(13,10)) 
sns.countplot(data    = df, 
              x       = "department", 
              palette = 'flare',
              hue     = 'recruitment_channel'
             )
plt.title('Perbandingan Perekrutan Karyawan Setiap Departemen Berdasarkan Channel Recruitment', fontsize = 15)
plt.xlabel('Departemen')
plt.ylabel('Count')
plt.show()
```

### Kesimpulan :
### Kesimpulan yang dapat diambil dari grafik di atas adalah terdapat data dari seluruh departemen dimana mayoritas karyawannya direkrut bukan berdasarkan sumber (mencari dari berbagai sumber) dan rekomendasi (rekomendasi dari karyawan lainnya) melainkan selain dari itu.

## ----------------------------------------------------------------
## 3. Grafik Boxplot

```python
plt.figure(figsize=(13,10))
sns.boxplot(data   = df,
            x      ='department',
            y      ='age',
)

plt.xlabel('Departemen')
plt.ylabel('Usia')
plt.title('Grafik Persebaran Data Usia Pada Setiap Departemen', fontsize = 15)
plt.show()
```

### Kesimpulan :
### Kesimpulan yang dapat diambil dari grafik di atas adalah rentang penyebaran usia karyawan pada masing-masing departemen adalah berkisar 24-40 tahun. 

## ----------------------------------------------------------------
## 4. Grafik Scatter Plot

```python
sns.set_style('whitegrid')
plt.figure(figsize=(13,7))

dataplot = df.sort_values('avg_training_score')
sns.scatterplot(
    data   = dataplot,
    x      = 'department',
    y      = 'avg_training_score',
    hue    = 'age',
    color  = '#F2789F'
    )

plt.xlabel('Departemen')
plt.ylabel('Rata-rata skor training')
plt.title('Grafik Perbandingan Data Karyawan Pada Setiap Rata-rata Skor Training Berdasarkan Usia', fontsize = 15)
plt.show()
```

### Kesimpulan :
### Kesimpulan yang dapat diambil dari grafik di atas adalah terdapat data rata-rata usia karyawan yang diterima pada tiap departemen adalah berusia dari rentang 24-40 tahun dengan berdasarkan rata-rata skor training.
